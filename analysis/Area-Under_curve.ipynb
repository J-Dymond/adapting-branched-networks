```python
normalised_depth = np.linspace(0,1,19)
layerwise_depth = np.arange(0,unbranched_mean.shape[0])
baseline_auc = np.trapz(unbranched_mean,normalised_depth)

def get_label(branch_weights):
    string = "("
    for weight in branch_weights:
        weight = np.round(weight,decimals=3)
        string = string + str(weight) + " : "
    string = string[:-3]+')'
    return(string)

def get_figs(means,branches,prefix="",model_type=""):
    auc=[]
    labels = []

    fig, ax1 = plt.subplots()
    fig, ax2 = plt.subplots()
    plt.title("Layer-wise accuracy improvement", size=14)
    plt.xlabel("Layer-wise Model Depth", size=13)
    plt.ylabel("Accuracy Difference", size=13)
    plt.xticks(np.arange(0,unbranched_mean.shape[0]+1))

    fig, ax3 = plt.subplots()
    plt.title("Layer-wise cumulative accuracy improvement", size=14)
    plt.xlabel("Layer-wise Model Depth", size=13)
    plt.ylabel("Cumulative Accuracy Difference", size=13)
    plt.xticks(np.arange(0,unbranched_mean.shape[0]+1))


    ax1.plot(normalised_depth,unbranched_mean)

    for i in range(len(means)):
        mean = means[i]
        ax1.plot(normalised_depth,mean)
        auc.append(np.trapz(mean,normalised_depth))
        diff = mean-unbranched_mean
        cum_sum = np.cumsum(diff)
        
        name = (model_type + get_label(branches[i]))
        labels.append(name)
        
        ax2.plot(layerwise_depth,diff,label=name)
        ax3.plot(layerwise_depth,cum_sum,label=name)

    ax2.axhline(y=0,ls='-',alpha=0.3,color='maroon',label="Unbranched baseline")
    ax2.axvline(x=5,ls='--',ymin=0.05,ymax=0.95,alpha=0.4,color='grey')
    ax2.axvline(x=9,ls='--',ymin=0.05,ymax=0.95,alpha=0.4,color='grey')
    ax2.axvline(x=13,ls='--',ymin=0.05,ymax=0.95,alpha=0.4,color='grey')
    ax2.legend(title='Weightings- (Exit 1 : Exit 2 : Exit 3 : Final Exit)',loc=1)
    ax2.set_xticks(np.arange(0,unbranched_mean.shape[0]+1))
    
    ax3.axhline(y=0,ls='-',alpha=0.3,color='maroon',label="Unbranched baseline")
    ax3.axvline(x=5,ls='--',ymin=0.05,ymax=0.95,alpha=0.4,color='grey')
    ax3.axvline(x=9,ls='--',ymin=0.05,ymax=0.95,alpha=0.4,color='grey')
    ax3.axvline(x=13,ls='--',ymin=0.05,ymax=0.95,alpha=0.4,color='grey')
    ax3.legend(title='Weightings- (Exit 1 : Exit 2 : Exit 3 : Final Exit)',loc=4)
    ax3.set_xticks(np.arange(0,unbranched_mean.shape[0]+1))
    
    ax2.figure.savefig(save_directory+prefix+'acc_improvement.pdf', bbox_inches = 'tight')
    ax3.figure.savefig(save_directory+prefix+'cum_acc_improvement.pdf', bbox_inches = 'tight')
    
    return np.array(auc),labels
```
